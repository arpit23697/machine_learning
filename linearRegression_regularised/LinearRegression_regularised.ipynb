```python
alpha = 0.03
```

```python
epsilon = 0.0000001
```

```python
lambd = [0.0 ,0.03 , 0.1 ,0.3 , 1.0 , 3.0 , 10 , 30 , 100]
```

```python
def compute_J (X_compute , y_compute , theta , lambd):
    m = X_compute.shape[0]
    
    error = (X_compute.dot(theta) - y_compute).transpose()
    error = error * error
    
    error2 = ((theta*theta)[1 , :]).sum()
    error2 = error2 * lambd / (2*m)
    
    error = (error.sum()) / (m * 2)
    error = error + error2
    return error
```

```python
def gradient (X_compute , y_compute , theta , alpha , lambd):
    m = X_compute.shape[0]
    n = theta.shape[0]
    temp = (X_compute.dot(theta) - y_compute).transpose()
    grad = (temp.dot(X_compute) / m_train ).transpose()
    
    temp2 = theta[1 , :].sum()
    temp2 = temp2 * lambd / (m)
    
    grad = grad + temp2
    return grad
```

```python
def compute_cost_test_cv (X_compute , y_compute , theta):
    m = X_compute.shape[0]
    
    error = (X_compute.dot(theta) - y_compute).transpose()
    error = error * error    
    error = (error.sum()) / (m * 2)
    return error
```

```python
#this is for the gradient descent block
def gradient_descent(X , y , lambd):
    n = 2                      #this one have only two features
    num_of_iteration = 0
    theta = np.ones((n , 1))
    cost1 = compute_J (X , y , theta , lambd)


    J = np.ones((1 , 2))
    while (True):
        num_of_iteration+=1

        #this is to update the value of theta
        grad = gradient(X , y , theta,alpha , lambd)
        theta = theta - alpha * grad

        #this is to compute the error and terminate gradient descent when the J almost becomes constant
        cost2 = compute_J (X , y ,theta , lambd)
        track = np.array([num_of_iteration , cost2])
        J = np.vstack((J , track))

        if (abs(cost2 - cost1) < epsilon):
            break
        cost1 = cost2
    return (J , num_of_iteration , theta)
```

```python
analysis = np.ones( (1 , 2)) 
analysis_train = np.ones((1 , 2))

final_lambda = lambd[0]
min_cost = math.inf
index = 0
for l in lambd:
    ( J , num_of_iteration , theta) = gradient_descent(X , y , l)
    cost = compute_cost_test_cv (X_cv , y_cv , theta)
    cost_train = compute_cost_test_cv (X , y , theta)
    
    index+=1
    
    analysis = np.vstack((analysis , np.array([index , cost])))
    analysis_train = np.vstack((analysis_train , np.array([index , cost_train])))
    if (min_cost > cost):
        min_cost =cost
        final_lambda = l
    

print(final_lambda , min_cost)
```

```python
plt.xlabel("lambda")
plt.ylabel("cost")
plt.title("cost Vs lambda")
plt.plot(analysis[1: , 0] , analysis[1: , 1] , label = "cross_validation")
plt.plot(analysis_train[1: , 0] , analysis_train[1: , 1]  , label = "training")
plt.legend()
```

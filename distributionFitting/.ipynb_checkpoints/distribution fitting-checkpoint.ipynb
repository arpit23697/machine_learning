```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import factorial
```

```python
file = open ("randomsample.txt" , "r")
X = [int(x) for x in file.readline().strip().split(',')]
```

```python
print("Number of data points : " , len(X))
print("Maximum value " ,  max(X))
print("Minimum value " , min(X))
```

```python
X = np.asarray(X)
```

```python
print("Shape of X is " , X.shape)
print("Minimum of X :" , X.min() )
print("Maximum of X : " , X.max())
```

```python
freq = np.bincount(X)
print("Printing the frequency of each element")
for i , j in zip (range(9) , freq):
    print(i , " - " , j)
```

```python
#this is to plot the histogram
plt.figure(figsize = (6 , 5) )
plt.hist(X , bins = np.arange(X.min() , X.max() + 2) - 0.5, histtype = 'stepfilled' , align = "left")
plt.title("Visualisation")
plt.xlabel("Value of data points")
plt.ylabel("Frequency")
```

```python
#this is to fit the gaussian model
n = X.shape[0]
mean = X.sum() / n
var = ((mean - X)**2).sum()
var = var / (n - 1)

print("Mean of the data point : " , mean)
print("var of the data point is  : " , var)

```

```python
line = np.linspace (-0.5 , 8.5 , 100)
print(line.shape)
#generating the gaussian function
gauss = (((line - mean)**2) / (2*var))
gauss = np.exp(-gauss)
gauss = (1/ np.sqrt(2 * np.pi * var))*gauss
```

```python
#plotting with the bin
#this is to plot the histogram
plt.figure(figsize = (6 , 5) )
plt.hist(X , bins = np.arange(X.min() , X.max() + 2) - 0.5 , density = True  , histtype = 'stepfilled')
plt.title("Visualisation and gaussian")
plt.xlabel("data points")
plt.ylabel("Gaussian and frequenct")
plt.plot(line , gauss , label = "gaussian")
plt.legend()
```

```python
#this is to plot the poisson function that fits the curve and then comparing it with the gaussian distribution
lamb = X.sum() / n
line2 = np.linspace(0 , 8.5 , 100)
y = np.exp(-lamb)*np.power(lamb, line2)/factorial(line2)

plt.figure(figsize= (8 , 6))
plt.hist(X , bins = np.arange(X.min() , X.max() + 2) - 0.5 , density = True  , histtype = 'stepfilled')
plt.plot(line , y , label = "Poisson distribution")
plt.plot(line , gauss , label = "gaussian")
plt.xlabel("Value of x")
plt.ylabel("Probability")
plt.legend()

```

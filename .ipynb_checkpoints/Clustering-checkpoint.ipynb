```python
#this is for importing the libraries
import numpy as np
import matplotlib.pyplot as plt

#sklearn libraries
from sklearn.datasets import make_moons
from sklearn.datasets import make_circles
from sklearn.cluster import KMeans
from sklearn.cluster import SpectralClustering
from sklearn.cluster import DBSCAN
from sklearn.cluster import Birch
from sklearn.cluster import AgglomerativeClustering
```

```python
#function for plotting the data set
def discrete_scatter (X , y , mark , classes, markerWidth = 1 , markerSize = 60 ):
    colors = ['red' , 'yellow', 'blue' , 'green' , 'purple' , 'brown' , 'black' , 'gray' , 'olive' , 'cyan']
    for i in range(classes):
        temp = X[y==i]
        plt.scatter(temp[: , 0] , temp[:, 1] , label = "Class " + str(i) , marker = mark,
                   linewidth = markerWidth , edgecolor = 'black' , s = markerSize , color = colors[i])
    
    plt.xlabel("Feature1")
    plt.ylabel("Feature2")
    plt.legend()
    plt.title("Data points")
```

```python
#to plot the discrete data points
def discrete_scatter2 (X , y , mark , classes , markwidth, marksize , name = "Data points"):
    colors = ['white' , 'red' , 'yellow', 'blue' , 'green' , 'purple' , 'brown' , 'black' , 'gray' , 'olive' ,
              'cyan' , 'gray' , 'silver' , 'whitesmoke' , 'aliceblue' , 'slategray']
    for i in range(-1 , classes):
        temp = X[y == i]
        plt.scatter(temp[: , 0 ] , temp[: , 1] , s = marksize , linewidth = markwidth , 
                   edgecolor= 'black' , label = "Class " + str(i) , marker = mark , color = colors[i+1])
    plt.xlabel ("Feature1")
    plt.ylabel ("Feature2")
    plt.legend ()
    plt.title (name)
```

```python
#original data set 1

X1 , y1 = make_moons (n_samples=300 , noise = 0.05 , random_state=0)
plt.figure(figsize= (8 , 6))
discrete_scatter (X1 , y1, 'o' , 2 , 1 , 100)
```

```python
#second original dataset
X2 , y2 = make_circles (n_samples=400 , noise = 0.05 , random_state=0 , factor = 0.5)
plt.figure(figsize= (8 , 6))
discrete_scatter (X2 , y2 , 'o' , 2 , 1, 100)
```

```python
#this is the function to create dataset with the name
#given the coordinate it will create the data set in between these two
def create_line (x1 , y1 , x2 , y2 , shape , n_sample = 100):
    if (shape == "horizontal"):
        temp1 = x1 + np.random.rand(n_sample, 1) * (x2 - x1)
        temp2 = (y1+y2)/2  + np.random.randn (n_sample , 1)* 0.1
        return np.hstack ([temp1 , temp2])
    
    if (shape == "vertical"):
        temp1 = y1 + np.random.rand(n_sample , 1) * (y2 - y1)
        temp2 = x1 + np.random.randn (n_sample , 1) * 0.1
        return np.hstack ([temp2 , temp1])
    
    if (shape == "slant1"):
        temp1 = x1 + np.random.rand(n_sample, 1) * (x2 - x1)
        temp2 = (y1 - x1) + temp1 + np.random.randn(n_sample , 1) * 0.1
        return np.hstack ([temp1 , temp2])

    if (shape == "slant2"):
        temp1 = x1 + np.random.rand(n_sample, 1) * (x2 - x1)
        temp2 = -temp1 + (y1 + x1)+ np.random.randn(n_sample , 1) * 0.1       
        return np.hstack ([temp1 , temp2])
    
    return False
```

```python
def create_circle (x , y , radius , shape , n_sample = 100):
    if (shape == "right"):
        r =  radius + np.random.randn(n_sample , 1 ) * 0.1
        theta = -np.pi/2 + np.random.rand(n_sample , 1) * np.pi
        
        print (np.max (theta) , np.min (theta))
        temp1 = x + r * np.cos (theta)
        temp2 = y + r * np.sin (theta)
        
        return np.hstack ([temp1 , temp2])

    if (shape == "left"):
        r =  radius + np.random.randn(n_sample , 1 ) * 0.1
        theta = np.pi/2 + np.random.rand(n_sample , 1) * np.pi
        temp1 = x + r * np.cos (theta)
        temp2 = y + r * np.sin (theta)
        
        return np.hstack ([temp1 , temp2])
```

```python
iit = create_line (0 , 0 , 0 , 5 , "vertical" , 50)
iit = np.vstack ([iit , create_line (-2.5 , 0 , 2.5 , 0 , "horizontal" , 50)])
iit = np.vstack ([iit , create_line (-2.5 , 5 , 2.5 , 5 , "horizontal" , 50)])
iit = np.vstack ([iit , create_line (7.5 , 0 , 7.5 , 5 , "vertical" , 50)])
iit = np.vstack ([iit , create_line (5 , 0 , 10 , 0 , "horizontal" , 50)])
iit = np.vstack ([iit , create_line (5 , 5 , 10 , 5 , "horizontal" , 50)])
iit = np.vstack ([iit , create_line (15 , 0 , 15 , 5 , "vertical" , 50)])
iit = np.vstack ([iit , create_line (12.5 , 5 , 17.5 , 5 , "horizontal" , 50)])

y = np.zeros ( (iit.shape[0] ) , dtype = np.int )
plt.figure(figsize= (8 , 6))
discrete_scatter (iit , y , 'o' , 1 , 1 , 100 )
```

```python
#create the data set with the name arpit
arpit = create_line (0 , 0 , 5 , 5 , "slant1" , 100)
arpit = np.vstack ([arpit , create_line (5 , 5 , 10 , 0 , "slant2" , 100)])
arpit = np.vstack ([arpit , create_line (2.5 , 2.5 , 7.5 , 2.5 , "horizontal" , 50)])

arpit = np.vstack ([arpit , create_line (12 , 0 , 12 , 5 , "vertical" , 100)])
arpit = np.vstack ([arpit , create_circle (12 , 3.75 , 1.25 , "right" , 50)])
arpit = np.vstack ([arpit , create_line (12 , 2.5 , 14.5 , 0 , "slant2" , 50)])

arpit = np.vstack ([arpit , create_line (17 , 0 , 17 , 5 , "vertical" , 100)])
arpit = np.vstack ([arpit , create_circle (17 , 3.75 , 1.25 , "right" , 50)])

arpit = np.vstack ([arpit , create_line (23 , 0 , 23 , 5 , "vertical" , 100)])
arpit = np.vstack ([arpit , create_line (21 , 0 , 25 , 0 , "horizontal" , 50)])
arpit = np.vstack ([arpit , create_line (21 , 5 , 25 , 5 , "horizontal" , 50)])

arpit = np.vstack ([arpit , create_line (28 , 0 , 28 , 5 , "vertical" , 100)])
arpit = np.vstack ([arpit , create_line (26 , 5 , 30 , 5 , "horizontal" , 50)])


plt.figure(figsize= (8 , 6))
discrete_scatter (arpit , np.zeros ((arpit.shape[0] ) , dtype = np.int) , 'o' , 1 , 1 ,100)
plt.axes().set_aspect('equal', 'datalim')
```

```python
#on half moon
cluster1_kmeans = KMeans(n_clusters=2)
cluster1_kmeans.fit(X1)

plt.figure(figsize= (8 , 6))
discrete_scatter (X1 , cluster1_kmeans.labels_ ,'o' ,2 , 1, 100)
discrete_scatter(cluster1_kmeans.cluster_centers_ , np.array([0 , 1]) , '^' , 2 , 3, 200)
```

```python
#on half moon
cluster2_kmeans = KMeans(n_clusters=2)
cluster2_kmeans.fit(X2)

plt.figure(figsize= (8 , 6))
discrete_scatter (X2 , cluster2_kmeans.labels_ ,'o' ,2 , 1, 100)
discrete_scatter(cluster2_kmeans.cluster_centers_ , np.array([0 , 1]) , '^' , 2 , 3, 200)
```

```python
#on half moon
cluster3_kmeans = KMeans(n_clusters=3)
cluster3_kmeans.fit(iit)

plt.figure(figsize= (8 , 6))
discrete_scatter (iit , cluster3_kmeans.labels_ ,'o' ,3 , 1, 100)
discrete_scatter(cluster3_kmeans.cluster_centers_ , np.array([0 , 1 , 3]) , '^' , 3 , 3, 200)
plt.axes().set_aspect('equal', 'datalim')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

```python
#on half moon
cluster4_kmeans = KMeans(n_clusters=5)
cluster4_kmeans.fit(arpit)

plt.figure(figsize= (8 , 6))
discrete_scatter (arpit , cluster4_kmeans.labels_ ,'o' ,5 , 1, 100)
discrete_scatter(cluster4_kmeans.cluster_centers_ , np.array([0 , 1 , 2, 3, 4]) , '^' , 5 , 3, 200)
plt.axes().set_aspect('equal', 'datalim')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

```python
models = [AgglomerativeClustering(n_clusters=5) ,
          Birch(threshold=0.5, branching_factor=50 , n_clusters=5),
            DBSCAN(eps=0.33),
          SpectralClustering(n_clusters=5),
         KMeans(n_clusters=5)]

model_name = ["Agglomerative clustering" , "Birch" , "DBSCAN" , "Spectral" , "KMeans"]

for name , model in zip (model_name , models):
    cluster4 = model.fit_predict(arpit)
    plt.figure(figsize= (8 , 6))
    discrete_scatter2 (arpit , cluster4 ,'o' ,cluster4.max() + 1 , 1 , 100 , name)
    plt.axes().set_aspect('equal', 'datalim')
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

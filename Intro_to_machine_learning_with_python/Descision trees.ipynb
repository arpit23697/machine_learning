```python
import mglearn
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#First , going to implement the descision tree for the classification problem
#Note if we are not going to do pruning , then the tree will overfit the model
#importing the dataset and splitting it
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
cancer = load_breast_cancer()
X_train , X_test , y_train , y_test = train_test_split(cancer.data , cancer.target)
```

```python
#the descision tree models are in DescisionTreeClassifier and DescisionTreeRegressor
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(random_state= 0)
#random state = 0, is used as the seed for the random number generator
#If no parameter for pruning is given , then the tree will be built until all the leaves are pure
#DecisionTreeClassifier and DecisionTreeRegressor support only the pre pruning
#two types of pruning ,prepruning before the tree is built , stop it using some criteria,
#post pruning , after the tree is built , then the nodes with less information are deleted
```

```python
#about the dataset
print("Shape of X_train is :" , X_train.shape)
print("Shape of X_test is : " , X_test.shape)
```

```python
#fitting the model
tree.fit(X_train , y_train)
```

```python
#printing the accuracies
print("Training accuracy : " , tree.score(X_train , y_train))
print("Test accuracy : " , tree.score(X_test, y_test))
#as expected the model is overfitting , so training accuracy is 1.0 while test accuracy is low
#it containg only the pure leaves
```

```python
#Now trying pre pruning
#now we limit the max_depth of the tree , so building the tree stops when the training data is not overfit
#so the model generalises well
tree2 = DecisionTreeClassifier(random_state = 0 , max_depth = 4)
tree2.fit(X_train , y_train)
```

```python
#printing the accuracies
print("Training accuracy :" , tree2.score(X_train , y_train) )
print("Test accuracy : " , tree.score(X_test , y_test))
#as expected the training accuracy is not 1.0 ; that means the model does not overfit the model
```

```python
#analysis
#we can visulaise the tree that is being made
from sklearn.tree import export_graphviz
export_graphviz(tree , out_file = "tree.dot" , class_names = ['malignant' , 'benign'] , 
               feature_names = cancer.feature_names , impurity = False , filled = True)

```

```python
#visualisinng the tree in the notebook itself
import graphviz
with open("tree.dot") as f:
    dot_graph = f.read()
graphviz.Source (dot_graph)
```

```python
#as can be seen without pruning the depth of the graph is 8
#plotting the other graph
export_graphviz(tree2 , out_file = "tree2.dot" , class_names = ['malignant' , 'benign'] ,
               feature_names = cancer.feature_names , impurity = False , filled = True)
with (open("tree2.dot")) as f:
    dot_graph = f.read()
graphviz.Source(dot_graph)

#here the maximum depth is 4
#OBSERVATIONS
#In each node , n_samples , give the number of samples that each node have
#In each node , values , gives the number of samples per class each node have
#each node uses only one feature to make the split
#color of the node is used to determine , which class is in the majority
```

```python
#feature importance of each feature in making the tree
#it tells which feature is most important in making the descisions
#0 means not used at all
#1 means the feature perfectly predicts the tree
#the feature importance of each feature always sum upto 1
#printing feature_importance of both the trees
print("Feature Importance: \n" , tree.feature_importances_)
```

```python
print("Feature Importance : \n" , tree.feature_importances_)
```

```python
#visualising the importance of each feature
def plot_feature_importances (model):
    n_features = cancer.data.shape[1]
    plt.barh (range(n_features) , model.feature_importances_ , align = 'center')
    plt.yticks(np.arange(n_features) , cancer.feature_names)
    plt.xlabel("Feature importance")
    plt.ylabel("Feature")
    plt.title('analysis')
```

```python
plt.figure(figsize = (10 , 7))
plot_feature_importances(tree2)
```

```python
plt.figure(figsize = (10 , 7))
plot_feature_importances(tree)
#if a feature have a low feature_importances , then it does not mean that this feature is uninformative
#it means the feature is not picked by the tree because another conveys the same information
#they do not encode , which class a feature is indicative of
```

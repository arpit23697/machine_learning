```python
import mglearn
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#It is another unsupervised learning algorithm , that aims to extract useful features
#It works similar to pca , and can be used for dimensionality reduction
#It is the same as pca , but here we want the component and coef to be non negative
#In this also , we try to write each data point as weighted sum of the components and the coefficients
#This method can only be applied , where each feature is non negative , as a non-negative sum of non-negative components
          #cannot be negative
#This is particularly useful for data that is created as a weighted sum of several independent components
#such as audio track of multiple people speaking, or music with many instruments
#Overall, can lead to a more interpretable components
#First thing to ensure , before applying NMF , is that the points are positive
#Non negative components are extracted as directions from (0 ,0) towards the data

#Now is the illustration of the NMF data set
rnd = np.random.RandomState(5)
X_ = rnd.normal(size = (300, 2))
X_blob = np.dot(X_ , rnd.normal(size = (2, 2 ))) + rnd.normal(size = 2) + 8


from sklearn.decomposition import NMF
nmf = NMF (random_state = 0)
nmf.fit(X_blob)
X_nmf = nmf.transform(X_blob)
print(nmf.components_)
```

```python
fig, axes = plt.subplots(1 , 3 , figsize = (15 , 8))
axes= axes.ravel()

axes[0].set_title("Original data with two component")
axes[0].scatter(X_blob[: , 0] , X_blob[: , 1] ,c = X_nmf[: , 0] ,cmap = 'viridis' , s = 60)
axes[0].arrow(0 , 0 , nmf.components_[0 , 0] ,nmf.components_[0 , 1] , width = 0.1 , head_width = 0.3 , color  = 'k' )
axes[0].arrow(0 , 0 , nmf.components_[1 , 0] ,nmf.components_[1 , 1] , width = 0.1 , head_width = 0.3 , color  = 'k' )
axes[0].set_aspect('equal')
axes[0].set_xlim(0 , X_blob[: , 0].max()+2)
axes[0].set_ylim(0 , X_blob[: , 1].max()+2)
axes[0].set_xlabel("Feature 1")
axes[0].set_ylabel("Feature 2")

nmf2 = NMF(n_components = 1)
nmf2.fit(X_blob)
X_nmf2 = nmf2.transform(X_blob)

axes[1].set_title("Original data with 1 component")
axes[1].scatter(X_blob[: , 0] , X_blob[: , 1] ,c = X_nmf[: , 0] ,cmap = 'viridis' , s = 60)
axes[1].arrow(0 , 0 , nmf2.components_[0 , 0] ,nmf2.components_[0 , 1] , width = 0.1 , head_width = 0.3 , color  = 'k' )
axes[1].set_aspect('equal')
axes[1].set_xlim(0 , X_blob[: , 0].max()+2)
axes[1].set_ylim(0 , X_blob[: , 1].max()+2)
axes[1].set_xlabel("Feature 1")
axes[1].set_ylabel("Feature 2")

axes[2].set_title('Data after dimensionality reduction')
X_inverse  = nmf2.inverse_transform(X_nmf2)
axes[2].scatter(X_inverse[: , 0] , X_inverse[: , 1] ,c = X_nmf[: , 0] ,cmap = 'viridis' , s = 60)
axes[2].set_xlim(0 , X_blob[: , 0].max()+2)
axes[2].set_ylim(0 , X_blob[: , 1].max()+2)
axes[2].set_aspect('equal')

#With nmf with two components , it is clear all the data points can be expressed as a linear comb of two comp
#If there are enough component to perfectly reconstruct the data , then the algo will choose directions ,that points towards the extreme
#If we only single component, then the component will point towards the mean of the data
#When reducing the number of components , entirely new data set is contructed
#Components in nmf are not ordered

#NMF uses a random initialiszation , so the result depend on the random seed
#For data set with less features , it does not really matter
```

```python
#applying the nmf to the labeled data set
#The main parameter is how many components we want to extract
#importing the data set
from sklearn.datasets import fetch_lfw_people
people = fetch_lfw_people(min_faces_per_person = 20 , resize = 0.7)
```

```python
print("The people keys are : \n" , people.keys())
print("The shape of people.data : ", people.data.shape)       #so there are 3023 images , with each image in each row
                                                        #the number of features is 5655 , which is equal to the number of pixels in each images
print("Shape of images is : " , people.images.shape)    # 87 x 65  is the dimension of the images
number_of_images = people.data.shape[0]
image_shape = (87 , 65)

print("Shape of target is : " , people.target.shape)
print("Shape of target names : " , people.target_names.shape) 
```

```python
#visualising some of the images
fig , axes = plt.subplots(2 , 5 , figsize = (15 , 8) , 
                         subplot_kw = {'xticks' : () , 'yticks' : ()})
for target , image , ax   in zip(people.target_names , people.images , axes.ravel()):
    ax.imshow(image , cmap = 'gray')
    ax.set_title(target)
```

```python
#making the data set better
mask = np.zeros(people.target.shape, dtype = bool)
for target in np.unique(people.target):
    mask[np.where(people.target == target)[0][:50]] = 1
    
X_people = people.data[mask]
y_target = people.target[mask]

X_people = X_people / 255
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X_people , y_target)
```

```python
#plotting how to recontruct image from the components
fig , axes = plt.subplots(3 , 5 , figsize = (15 , 12) , 
                         subplot_kw = {'xticks' : () , 'yticks' : ()})

for i , image ,ax in zip(range(3) , X_train , axes[: , 0]):
    ax.imshow(image.reshape(image_shape), cmap = 'gray')
    ax.set_title('Original image')

for n_component , i in zip([10 , 50 , 100 , 500] , range(4)):
    print(n_component , end = "  ")
    nmf = NMF(n_components = n_component , random_state = 0)
    nmf.fit(X_train)
    X_change = nmf.transform(X_train)
    X_inverse = nmf.inverse_transform(X_change)
    print('Done')
    for j , image , ax in zip(range(3) , X_inverse , axes[: , i+1]):
        ax.imshow(image.reshape(image_shape) , cmap = 'gray')
        ax.set_title("components " + str(n_component))
```

```python
#the quality of the back transformed data is similar to when using PCA , slightly worse, as expected as
#PCA selects the best direction in the data set
#visualising the components of the nmf


nmf = NMF(n_components = 15 , random_state = 0)
nmf.fit(X_train)
X_train_transform = nmf.transform(X_train)
X_test_transform = nmf.transform(X_test)

fig , axes = plt.subplots(3,  5 , figsize = (15 , 12) , 
                         subplot_kw = {'xticks' : () , 'yticks' : ()})

for i , image , ax in zip(range(15) , nmf.components_ , axes.ravel()):
    ax.imshow(image.reshape(image_shape) , cmap = 'gray')
    ax.set_title("component " + str(i))
    
#these resemble the prototype of faces more closely
#component 4 , face rotated to the left
#component 5 , face rotated to the right
```

```python
#dealing with the signals
S = mglearn.datasets.make_signals()
print(S.shape)

#plotting the graph to visualise it
plt.figure(figsize = (12 , 2))
plt.plot(S , '-')
plt.xlabel('Time')
plt.ylabel('magnitude')
```

```python
#Lets say we have 100 different divices for the measurement
#We want to recover the decomposition of the mixed signal into the original components
#mix data into a 100 dimensional space
A = np.random.RandomState(0).uniform(size = (100, 3))
X = np.dot(S , A.T)
print("Shape of measurement : " , X.shape)
```

```python
#recovering the components using nmf
nmf = NMF(n_components = 3 , random_state = 42)
nmf.fit(X)
S_ = nmf.transform(X)
print("Recovered signal shape : " , S_.shape)
```

```python
#for comparison also apply the PCA 
from sklearn.decomposition import PCA
pca = PCA(n_components = 3)
H = pca.fit_transform(X)


```

```python
#plotting the curves
models = [X , S , S_ , H]
names = ['Observations "First thres components"' , 'True sources' , "NMF recovered signal" , "PCA recovered signals"]

fig ,axes = plt.subplots(4 , 1, figsize = (12 , 10) , subplot_kw = {'xticks' : () , 'yticks' : ()})
for model , name , ax in zip (models , names , axes.ravel()):
    ax.set_title(name)
    ax.plot(model[: , :3] , '-')
    
#it is clear that the signal recovery be nmf is much better as compared to the pca data
```

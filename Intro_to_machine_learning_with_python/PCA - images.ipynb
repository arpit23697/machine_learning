```python
import mglearn
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Here we are going to deal withfeature extraction
#The idea of feature extraction is that it is possible to find a represntation of the data , that is better suited for 
#analysis then the raw data

#importing the dataset
from sklearn.datasets import fetch_lfw_people
people = fetch_lfw_people(min_faces_per_person = 20 , resize = 0.7)
```

```python
print("The people keys are : \n" , people.keys())
print("The shape of people.data : ", people.data.shape)       #so there are 3023 images , with each image in each row
                                                        #the number of features is 5655 , which is equal to the number of pixels in each images
print("Shape of images is : " , people.images.shape)    # 87 x 65  is the dimension of the images
number_of_images = people.data.shape[0]
image_shape = (87 , 65)

print("Shape of target is : " , people.target.shape)
print("Shape of target names : " , people.target_names.shape)      # so total number of labels is 62
```

```python
#visualising some of the pictures
fix, axes = plt.subplots (2 , 5 , figsize = (15 , 8) , subplot_kw = {'xticks' : () , 'yticks' : ()})
for target , image , ax in zip (people.target , people.images , axes.ravel()):
    ax.imshow(image , cmap = 'gray')
    ax.set_title(people.target_names[target])
```

```python
#analysing the data
counts = np.bincount(people.target)
for count , name in zip (counts , people.target_names):
    print(name + " : " + str(count))
    
#you can see the data is skewed having 530 images of george bush
```

```python
#so we need to correct the dataset
#So we make only upto 50 images of each person
mask = np.zeros (people.target.shape , dtype = np.bool)
for target in np.unique(people.target):                    #np.unique returns the unique image sorted order
    mask[np.where(people.target == target)[0][:50]] = 1    #takes a maximum of 50 images of each person
X_people = people.data[mask]
y_people = people.target[mask]

X_people = X_people / 255
```

```python
#print shape of X_people and y_people
print("Shape of X_people is : " , X_people.shape)
print("Shape of y_people is : " , y_people.shape)
```

```python
#building a classifier where each person is separate class
#Since there are very few images so it is hard to train the classifier
#A simple solution would be to use the knn classifier with the value of 1
#It will pick the class with the nearest images

#increasing the number of neighbors will only make the model weaker
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X_people , y_people)

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 1)
knn.fit(X_train , y_train)
print("Test set accuracy : " , knn.score(X_test , y_test))
```

```python
#computing distances in the original pixcel image is a very bad thing
#so the better way is to use the PCA vector space
#using distances along the pca , can improve the accuracy
#enable the whitening option of PCA , which help the pca to have the same range as the original one
#this is like using the standard scaler after the transformation
#whitening not only correspong to rotating the data, but also scaling it to the the same scale
#this is done so that all the features haev the same scale
from sklearn.decomposition import PCA
pca = PCA(n_components = 100 , whiten = 100 , random_state = 0 )
pca.fit(X_train)
X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)

#shape of X_train_pca and X_test_pca
print("Shape of X_train_pca : " , X_train_pca.shape)
print("Shape of X_test_pca : " , X_test_pca.shape)
#The new data have only 100 features ; the starting 100 principal component
```

```python
#fitting the model
knn = KNeighborsClassifier(n_neighbors = 1)
knn.fit(X_train_pca , y_train)
print("Test set accuracy is : " , knn.score(X_test_pca , y_test))

#as expected the accuracy is increased
#so as expected , PCA is better representation 
```

```python
print("Pca.components_.shape : " , pca.components_.shape)
#all the 100 components , each in each row 
#we can visualize the starting components
#visualising the starting components 
fig , axes = plt.subplots(3 , 5 , figsize = (15 , 12) , 
                         subplot_kw = {'xticks' : () , 'yticks' : ()})
for i , ax , components in zip (range(15) ,axes.ravel() , pca.components_):
    ax.imshow(components.reshape(image_shape) , cmap = 'gray')
    ax.set_title("compoent " + str(i))
    
#So the way the human percieve an image is different from the one , when machine percieve the image
```

```python
#with the increase in the number of principal component , the reconstruction of the image
#using more components result in better image orientation
#if number of components = number of features in orignal space , we donot lose info and reconstruct the image
fig , axes = plt.subplots (3, 5 , figsize = (15 , 12) ,
                          subplot_kw = { 'xticks' : ()  , 'yticks' : ()})
for ax , i , image in zip (axes[: , 0] , range(3) , X_people ):
    ax.imshow(image.reshape(image_shape) , cmap = 'gray')
    ax.set_title("Original image")
    
for i , n_component in zip (range(1 , 6) , [10 , 50 , 100 , 500]) :
    pca = PCA (n_components = n_component , whiten = 100)
    pca.fit(X_people)
    X_people_changed = pca.transform(X_people)
    X_inverse = pca.inverse_transform(X_people_changed)
    
    for ax , j , image in zip (axes[: , i] , range(3) , X_inverse):
        ax.imshow(image.reshape(image_shape) , cmap = 'gray')
        ax.set_title(str(n_component) + " components")
```

```python
#Using the pca to visualise the images in the 2D plot
pca = PCA(n_components = 2)
pca.fit(X_people)
X_transformed = pca.transform(X_people)
```

```python
plt.figure (figsize = (8  , 6))
plt.title('2D plot for face data set')
for i in range(62):
    X_temp = X_transformed[y_people == i]
    plt.scatter(X_temp[: , 0] , X_temp[: , 1])
plt.xlabel("First principal component")
plt.ylabel("Second principal component")
#no separation of the classes visible
```

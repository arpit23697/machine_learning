```python
import mglearn
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#they refer to a collection of clustering algorithms , that all bulid upon the same principle
#The algo starts by declaring each point its own cluster and then merges the two most similar clusters until some stopping
#criteria is satisfied
#The stopping criteria in scikit learn is the number of clusters
#Some of the linkage criteria for determining two most similar clusters are
#   ward - It merges two clusters such that the variance between the two clusters increase the least
#   average - merges two clusters that have the smallest average distance bewtween them
#   maximum - merges two clusters that have the smallest maximum distance between their points

#ward works good and it is the default
#If the clusters have very different sizes , maximum and average are going to work better
```

```python
def discrete_scatter (X , y , mark , classes , markwidth = 1 , marksize = 60):
    colors = ['red' , 'yellow', 'blue' , 'green' , 'purple' , 'brown' , 'black' , 'gray' , 'olive' , 'cyan']
    for i in range(classes):
        temp = X[y == i]
        plt.scatter(temp[: , 0 ] , temp[: , 1] , s = marksize , linewidth = markwidth , 
                   edgecolor= 'black' , label = "Class " + str(i) , marker = mark , color = colors[i])
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
```

```python
#importing the data set and applying the algomerative clustering algorithm
from sklearn.datasets import make_blobs
X , y = make_blobs(random_state = 1)

#plotting the graph
plt.figure(figsize = (8 , 6))
discrete_scatter (X , y,  'o' , 3 , 1 , 100)
plt.title("Original data points")
```

```python
#using the algomerative clustering algo with 2 clusters
from sklearn.cluster import AgglomerativeClustering
agg = AgglomerativeClustering(n_clusters = 3)
assignment = agg.fit_predict(X)                    #it only have the fit_predict method. it is not going to apply to the new data

plt.figure(figsize = (8 , 6))
discrete_scatter (X , assignment , 'o' , 3 , 1, 100)
plt.title("Agglomerative algorithm")

#As expected the agglomerative algorithm is working perfectly for this dataset
```

```python
#for choosing the right number of clusters you can plot the dendograms ,as they are the hierarichal clustering
#Every point in this cluster makes a journey from being a single point cluster to some final cluster
#scikit learn do not have the functionality to bulit dendograms
#We are going to use scipy to built the dendograms
X , y = make_blobs(n_samples = 12 , random_state = 0)

from scipy.cluster.hierarchy import dendrogram , ward
linkage_array = ward(X)


#this is to plot the dendrogram
plt.figure(figsize = (8 , 6))
dendrogram (linkage_array)

#plotting the lines for the threshold
bounds = np.linspace (0 , 150 , 100)
plt.plot(bounds  , np.ones(bounds.shape) * 7.25, '--k' )
plt.plot(bounds  , np.ones(bounds.shape) * 4, '--k' )

plt.text (123 , 7.25, 'Two Clusters' , fontdict = {'weight' : 'bold'})
plt.text (123 , 4, 'Three Clusters' , fontdict = {'weight' : 'bold'})
plt.title("Dendrogram")
plt.xlabel("Sample index")
plt.ylabel("Cluster distance")
#going from three to two clusters , is having a very large distance
#So it means it is better to divide the data into three clusters
```

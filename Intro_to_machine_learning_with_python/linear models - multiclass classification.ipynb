```python
#import mglearn
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#we are going to use the ove Vs rest method 
#In this we train the binary classifier for each of the classes , resulting in coef and intercept for each class
#Then we predict the class of the data point , by seeing for which class the prediction is highest
#importing the dataset
from sklearn.datasets import make_blobs
X , y = make_blobs(random_state = 42)
```

```python
#plotting the graph
def plot_data_points():
    n = 2
    class0 = np.ones((0 , n))
    class1 = np.ones((0 , n))
    class2 = np.ones((0 , n))
    for indices in range(X.shape[0]):
        if (y[indices] == 0):
            class0 = np.vstack((class0 , np.array([[X[indices , 0] , X[indices , 1]]])))
        elif (y[indices] == 1):
            class1 = np.vstack((class1 , np.array([[X[indices , 0] , X[indices , 1]]])))
        elif (y[indices] == 2):
            class2 = np.vstack((class2 , np.array([[X[indices , 0] , X[indices , 1]]])))

    plt.plot(class0[: , 0 ] , class0[: , 1] , 'bo' , label = 'class 0')
    plt.plot(class1[: , 0 ] , class1[: , 1] , 'go' , label = 'class 1')
    plt.plot(class2[: , 0 ] , class2[: , 1] , 'ro' , label = 'class 2')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
 
```

```python
#plotting the data points
plot_data_points()
plt.title('Data points')
plt.legend()
```

```python
#importing and training the model
from sklearn.svm import LinearSVC
model = LinearSVC().fit(X , y)
```

```python
#note the model uses the one vs rest approach
#since there are three classes , so the number of coef_ and intercept_ pair returned is 3
#To make prediction , see correspoding class ,the prediction is highest
print("Shape of coef_ : ", model.coef_.shape)
print("Shape of intercept_ : " , model.intercept_.shape)
```

```python
#plotting the graph correspoding to each class
line = np.linspace (-10 , 8 , 100)
plt.figure(figsize = (10 , 8))
plot_data_points()
for coef , intercept , color , labels  in zip (model.coef_ , model.intercept_ , ['b--' , 'g--' , 'r--'] 
                                               , ['class0 line' , 'class1 line' , 'class2 line']):
    plt.plot(line , -(coef[0]*line + intercept) / coef[1] , color , label = labels)
    
plt.legend(loc = 'lower right')
plt.title('Classification')
plt.ylim(-12 , 12)
#now all we have to do is to worry about the triangle in between 
#this is the triangle which all the three classes classifies as the rest
```

```python
#plotting the area of the classifier
line = np.linspace (-10 , 8 , 1000)
plt.figure(figsize = (10 , 8))
plot_data_points()
for coef , intercept , color , labels  in zip (model.coef_ , model.intercept_ , ['b--' , 'g--' , 'r--'] 
                                               , ['class0 line' , 'class1 line' , 'class2 line']):
    plt.plot(line , -(coef[0]*line + intercept) / coef[1] , color , label = labels)
    
plt.legend(loc = 'lower right')
plt.title('Classification Region')
plt.ylim(-12 , 12)

xx = line
yy = np.linspace (-12 , 12, 1000)
X1 , X2 = np.meshgrid(xx , yy)
X_grid  = np.c_[X1.ravel() , X2.ravel()]
descision_values = model.predict(X_grid)

plt.contourf(X1 , X2 , descision_values.reshape(X1.shape) ,alpha = 0.5)

```

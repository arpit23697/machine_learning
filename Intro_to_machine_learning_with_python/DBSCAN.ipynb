```python
import mglearn
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#to plot the discrete data points
def discrete_scatter (X , y , mark , classes , markwidth, marksize , ax):
    colors = ['white' , 'red' , 'yellow', 'blue' , 'green' , 'purple' , 'brown' , 'black' , 'gray' , 'olive' , 'cyan']
    for i in range(-1 , classes):
        temp = X[y == i]
        ax.scatter(temp[: , 0 ] , temp[: , 1] , s = marksize , linewidth = markwidth , 
                   edgecolor= 'black' , label = "Class " + str(i) , marker = mark , color = colors[i+1])
    return ax
```

```python
#It stands for density based spatial clustering of application with noise
#IT does not require the user to put the number of clusters a priori
#IT can even identify points that are not part of any cluster
#The idea DBSCAN , is that the clusters from dense regions of data, separated by regions that are relatively empty
#There are two parameter in DBSCAN - min_samples and eps
#If there are min_samples number of data points within a distance of eps , then the point is classified as eps
#At the end of the DBSCAN algo , there are three kinds of points
#core points , points that are within distance eps of core points (boundary points) and noise
#cluster membership of the boundary points depends on the order in which the points are visited
#Since there are fewer boundary points , so the order is not that important
```

```python
#applying DBSCAN
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
X , y = make_blobs(random_state = 0 , n_samples = 12)
dbscan = DBSCAN()
clusters = dbscan.fit_predict(X)
print("Cluster points : \n"  , clusters)
#Since all the points are classified as noise we need to increase the value of eps or decrease min_samples
```

```python
eps = [1.0 , 1.5 , 2.0 , 3.0]
min_samples = [2 , 3 , 5]

fig , axes = plt.subplots(3 , 4 , figsize= (15 , 15) , 
                         subplot_kw = {'xticks' : () , 'yticks' : ()} )


for i , e in zip(range(4) , eps):
    for j , sample in zip (range(3) , min_samples):
        dbscan = DBSCAN(eps = e , min_samples = sample)
        clusters = dbscan.fit_predict(X)
        discrete_scatter  (X , clusters, 'o' , y.max() + 1 , 1 , 50 , axes[j , i])
        discrete_scatter  (X[dbscan.core_sample_indices_] , clusters[dbscan.core_sample_indices_] , '^' , y.max() + 1 , 1 , 200 , axes[j , i])
        axes[j , i].set_title("eps = " + str(e) + "  ,  min_samples = " + str(sample) )

#triangles represent core samples
#small white circles represent noise
#small colored circles represent noise
#increasing means more points will be there in the cluster
#increasing min_samples mean fewer points will be the core points
#parameter eps is more important , as it somewhat tells what it means for the points to be close
#if you increase min_samples , then anythin that would be cluster with lesser points will not be cluster anymore
#min_sample = min_cluster_size
```

```python
#let's apply the DBSCAN algo on the two moon dataset
#before applying it would be better to scale the data
from sklearn.datasets import make_moons
X , y  = make_moons(n_samples = 200 , noise = 0.05 , random_state =0)

#scaling the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)

#using DBSCAN
dbscan = DBSCAN(eps = 0.2)
clusters = dbscan.fit_predict(X)
plt.figure(figsize = (8 , 6))
discrete_scatter(X , clusters , 'o' , clusters.max()+1 , 1 , 100 , plt.gca()) 
plt.title("DBSCAN algorithm on two moon data set with eps=0.2")
```

```python
import mglearn
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#PCA is a good approach for dimensionality reduction, to visualise your data with the help of scatter plot
#But the way it uses it , limits its usefullness , as we saw with the scatter plot of the faces
#There is a different class of algo for visualization , manifold algorith, that allow for more complex mappings
#One of the algorithm is the t-SNE
#Manifold algo are used for visualization , so they are rarely used to generate more than two features
#It compute a new representation of the training data , but do not allow to transform the new data
#They are used for visualisation only , and they are rarely used if the final goal is supervised learning
#The idea behind t-SNE is to find the 2D representation of the data , that preserves the distance between two points as best as possible
#It starts with a random 2D representation of the data points and then tries to make data points that are close closer
#and points that are far farther
#It focuses more on making the data points that are close ,then the ones that are far apart
#Using the t-SNE algorithm on the digit data set and seeing it for visualization
```

```python
#importing the dataset
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
#Description of the data set
print("Key of the data set are : \n" , digits.keys() )
print("Size of the data : " , digits.data.shape )      #so total number of data points are 1797
                                                         #total number of features are 64 whiich is the number of the pixels
print("Size of the targets : " , digits.target.shape)    
print("Number of classes : " , digits.target_names.shape)
print("The target names are : \n" ,digits.target_names )        #all of them are integer

print("Size of image is : " , digits.images.shape)
print("Shape of one image : " , digits.images[0].shape)
image_shape = digits.images[0].shape
```

```python
#printing the images
fig , axes = plt.subplots(2 , 5 , figsize = (15 , 8) ,
                         subplot_kw = {'xticks' : () , 'yticks' : ()})
for ax , image in zip(axes.ravel() , digits.images):
    ax.imshow(image , cmap = 'gray')
```

```python
#applying pca on the data set ,and see how well it is going to be separated
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
pca.fit(digits.data)
X_pca  = pca.transform(digits.data)

print("shape of X_pca : " , X_pca.shape) 
```

```python
#plotting all the data points on the graph
plt.figure(figsize = (10 , 10))
plt.title("2D representation of the data")
plt.xlim(X_pca[: , 0].min() - 2 , X_pca[: , 0].max() + 2)
plt.ylim(X_pca[: , 1].min() - 2 , X_pca[: , 1].max() + 2)

colors = ['blue' , 'orange', 'green' , 'red' , 'purple' , 'brown' , 'black' , 'gray' , 'olive' , 'cyan']
for i in range(len(digits.data)):
    plt.text(X_pca[i , 0] , X_pca[i , 1] , str(digits.target[i]),
    color = colors[digits.target[i]] ,
            fontdict = {'weight' : 'bold'})
    
#as can be seen the data is very mixed up when using the pca algorithm
#digits 0 , 6 and 4 are relatively well separated
#most of the digits overlap significantly
```

```python
#applying the TSNE on the data set and comparing the results
#Since it does not support transforming the data , we have to use fit_transform

from sklearn.manifold import TSNE
tsne = TSNE(random_state = 42 )
X_tsne = tsne.fit_transform(digits.data)
```

```python
#plotting the graph
plt.figure(figsize = (10 , 10))
plt.title("2D representation of the data using TSNE")
plt.xlim(X_tsne[: , 0].min() - 2, X_tsne[: , 0].max() + 2)
plt.ylim(X_tsne[: , 1].min() - 2, X_tsne[: , 1].max() + 2)

colors = ['blue' , 'orange', 'green' , 'red' , 'purple' , 'brown' , 'black' , 'gray' , 'olive' , 'cyan']
for i in range(len(digits.data)):
    plt.text(X_tsne[i , 0] , X_tsne[i , 1] , str(digits.target[i]) , 
            color = colors[digits.target[i]] , 
            fontdict = {'weight' : 'bold'})
    
#this shows the data point are pretty well separated
```

```python
#importing the dataset
from sklearn.datasets import fetch_lfw_people
people = fetch_lfw_people(min_faces_per_person = 20 , resize = 0.7)
```

```python
pca = PCA(n_components = 2)
pca.fit(people.data)
X_pca = pca.transform(people.data)
```

```python
#plotting the graph
plt.figure(figsize = (10,10))
for i in range(62):
    temp = X_pca[people.target == i]
    plt.scatter(temp[: ,0 ],  temp[: , 1])
    
plt.title("Face data set with PCA")
plt.xlabel('First component')
plt.ylabel("Second component")

#the data seems to be very messy
```

```python
#using tsne
tsne = TSNE(n_components = 2)
X_tsne= tsne.fit_transform(people.data)
```

```python
plt.figure(figsize = (10 , 10))
for i in range(62):
    temp = X_tsne[people.target == i]
    plt.scatter(temp[: , 0] , temp[: ,1])
plt.title("Data with tsne")
#This is the data after tsne
```

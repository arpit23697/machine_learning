```python
import mglearn
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import make_blobs
X , y = make_blobs (centers = 4 , random_state = 8)
y = y%2
```

```python
#plotting the datap point
def plot_data_points(X , y):
    class0 = np.ones((0 , 2))
    class1 = np.ones((0 , 2))
    for indices in range(X.shape[0]):
        if (y[indices] == 0):
            class0 = np.vstack((class0 , np.array(X[indices])))
        else:
            class1 = np.vstack((class1 , np.array(X[indices])))
    plt.plot(class0[: , 0] , class0[: , 1] , 'o' , label = "class 0")
    plt.plot(class1[: , 0] , class1[: , 1] , 'r^' , label = "class 1")
    plt.xlabel('Feature 1')
    plt.ylabel("feature 2")
```

```python
def plot_decision_boundary(model):
    eps = 0.5
    x_min , x_max = X[: , 0].min() - eps , X[: , 0].max() + eps
    y_min , y_max = X[: , 1].min() - eps , X[: , 1].max() + eps
    xx = np.linspace (x_min , x_max , 100)
    yy = np.linspace (y_min , y_max , 100)

    X1 , X2 = np.meshgrid (xx , yy)
    X_grid = np.c_[X1.ravel() , X2.ravel()]

    Z = model.decision_function(X_grid).reshape(X1.shape)
    
    plot_data_points(X , y)
    plt.contourf(X1 , X2 , Z , alpha = 0.5 , levels = [Z.min() , 0 , Z.max()] )
   
```

```python
plot_data_points(X , y)
plt.title("Data points")
plt.legend()
```

```python
#a linear model for classification can only separate the points using the line
#So it will not be able to do a great job on this dataset
from sklearn.svm import LinearSVC
linear_svm = LinearSVC(C = 1.0).fit(X , y)
#you can improve the performance of this by tuning C
#larger C , more overfitting
#smaller C , means lesser overfitting
```

```python
#plotting the separator
plot_decision_boundary(linear_svm)
plt.title("Decision boundary")
plt.legend()
```

```python
#now add another feature which is the square of the another feature 
X_new = np.hstack((X , (X[: , 1] ** 2).reshape(X.shape[0] , 1) ))
print("Shape of X_new is : " , X_new.shape)
```

```python
#plotting the 3D graph to visualise the data points
from mpl_toolkits.mplot3d import Axes3D , axes3d
figure = plt.figure()
ax = Axes3D(figure , elev = -152 , azim = -26)
mask = y == 0
ax.scatter (X_new[mask , 0] , X_new[mask , 1] , X_new[mask , 2] , c = 'b')
ax.scatter(X_new[~mask , 0] , X_new[~mask , 1] , X_new[~mask , 2] , c = 'r' , marker = '^' )

#setting the labels
ax.set_xlabel("Feature 0")
ax.set_ylabel("Feature 1")
ax.set_zlabel("Feature1 ** 2")

#It is clear from the figure that it is possible to separate the data set now
```

```python
#fitting the model and plotting the separating plane
linear_svm_3d = LinearSVC().fit(X_new , y)
coef , intercept = linear_svm_3d.coef_.ravel() , linear_svm_3d.intercept_

figure = plt.figure()
ax = Axes3D(figure , elev = -152 , azim = -26)

x_min , x_max = X_new[: , 0].min() - 2 , X[: , 0].max() + 2
y_min , y_max = X_new[: , 1].min() - 2 , X[: , 1].max() + 2

xx = np.linspace(x_min , x_max , 50)
yy = np.linspace(y_min , y_max , 50)

XX , YY = np.meshgrid(xx , yy)
ZZ = (coef[0] * XX + coef[1] * YY + intercept) / -coef[2]

#rstride , cstride = 8 , means it for the row , col grids
ax.plot_surface(XX , YY , ZZ , alpha = 0.3)
mask = y == 0
ax.scatter (X_new[mask , 0] , X_new[mask , 1] , X_new[mask , 2] , c= 'b' )
ax.scatter (X_new[~mask , 0] , X_new[~mask , 1] , X_new[~mask, 2] , marker = '^' , c= 'r')

ax.set_xlabel("Feature 0")
ax.set_ylabel("Feature 1")
ax.set_xlabel("Feature1 ** 2")
```

```python
ZZ = YY ** 2
dec = linear_svm_3d.decision_function(np.c_[XX.ravel() , YY.ravel() , ZZ.ravel()]) #decision_function gives the distance of the sample X to the separating hyperplane
plt.contourf(XX , YY , dec.reshape(XX.shape) , alpha = 0.3 , levels = [dec.min() , 0, dec.max()])
plot_data_points(X , y)
plt.legend()
```

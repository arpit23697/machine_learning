```python
import mglearn
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#the way different models behave with different feature representation depends on the model itself
#Here we are going to work with wave dataset and two different models regression and tree
X , y = mglearn.datasets.make_wave (n_samples = 100)
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor

line = np.linspace (-3, 3 , 1000 , endpoint = False).reshape(1000 , 1)
plt.figure(figsize = (8 , 6))
plt.scatter (X , y , marker = 'o' , color = 'k')
#using the decision tree model
reg = DecisionTreeRegressor(min_samples_split = 3).fit(X , y)
plt.plot(line , reg.predict(line) , label = 'Decision tree')

#using the linear regression model
reg = LinearRegression().fit(X , y)
plt.plot(line , reg.predict(line) , label= 'Linear regression')

plt.xlabel("Feature")
plt.ylabel("Target")
plt.legend()
plt.title("Comparison")
```

```python
#So here we see that the linear models do not work that much great because they can only fit a line
#One way to improve this is to use binning of the model to split it into multiple features
#We are going to partition -3 to 3 into 10 bins
#A data point is represented by which bin it fall into
bins = np.linspace (-3 , 3 ,11)
print(bins)
#first bin contain all the data point from -3 to -2.4
#second bin contain all the data point from -2.4 to -1.8
```

```python
which_bin = np.digitize(X , bins = bins)
print("Data points : \n" , X[:5] )
print("Which bin : \n" , which_bin[:5])
#here we converted the one continuous feature into one categorical feature which mention which the data is into
```

```python
#creating the one hot encoding of the bin data set
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse = False)

encoder.fit(which_bin)
X_binned = encoder.transform(which_bin)
print("Few rows of the X_binned is : \n" , X_binned[:5])
print("Shape of X_binned is  : \n" , X_binned.shape)
#this is the example to create one hot encoding using the bin data set
#the same you can do with the pandas get_dummies function
```

```python
line_binned = encoder.transform(np.digitize(line , bins = bins))
plt.figure(figsize = (8 , 6))
reg = LinearRegression().fit(X_binned , y)
plt.plot(line , reg.predict(line_binned) , 'g' , label = "Linear Regression model")

reg = DecisionTreeRegressor(min_samples_split = 3).fit(X_binned , y)
plt.plot(line , reg.predict(line_binned) , 'r--', label = "Decision tree model" )

plt.scatter(X , y, color = 'k' , marker = 'o')
plt.vlines (bins , -3 , 3 , linewidth = 1 , alpha = .2 )

plt.xlabel("Feature points")
plt.ylabel("Target")
plt.legend()
plt.title('Comparison after binning')

#the model for both linear regression and the decision tree is the same
#Introducing the binning add linear model more typical
#But it is not very useful for the decision tree
#so it is better not to use binning with decision tree
#the lines in this are not going to have any slope because there is no info about the x
#We only know in which bin the point lies in , we donot know where on the x-axis point lies
```

```python
#Now deal with the interaction and the polynomials
#in order to add the slope to this data set we can add the original feature back in
#But this will result in only one slope that is going to be shared across all the bins
X_combined = np.hstack([X , X_binned])
print("Shape of X_combined is : " , X_combined.shape)
```

```python
#now plotting the regression model
reg = LinearRegression().fit(X_combined , y)

plt.figure(figsize = (8 , 6))

line_combined = np.hstack([line , line_binned])
plt.plot(line , reg.predict(line_combined) , '--r')

plt.scatter(X , y , color = 'k' , marker = 'o')
plt.vlines (bins , -3 , 3, alpha = 0.2 , linewidth=1)

plt.xlabel("Features")
plt.ylabel("Target")
plt.title("Comparison with interaction and binning")

#the slope that is learned is shared across all the bins
#Since the slope is shared across all the bins it do not seem to be that useful
#we can include this by adding another set of features which indicate in which the point is and where it lies on the x_axis
```

```python
#now to have different slopes across different models
X_product = np.hstack([X_binned , X * X_binned])
print("Shape of X_product is : " , X_product.shape)
print("Starting 5 rows are : \n" , X_product[:5])
```

```python
#fitting the model and seeing the result
#Now the model not only the bins but also where on x_axis the point lies
lines_product = np.hstack([line_binned , line*line_binned])
reg = LinearRegression().fit(X_product , y)

plt.figure(figsize = (8 , 6))

plt.plot(line , reg.predict(lines_product) , '--r')
plt.scatter(X, y, color = 'k' , marker= 'o')

plt.vlines(bins , -3 , 3, alpha = 0.2, linewidth = 1)
plt.xlabel("Feature")
plt.ylabel("Target")
plt.title("Binned and interaction")
```

```python
#Another way to expand continuous features is to use polynomial features
#you can add the polynomial features one by one. But the best way is to use scikit library
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree = 10 , include_bias = False)
#include_bias = False means that the polynomials with power 0 are not included
poly.fit(X)
X_poly = poly.transform(X)
```

```python
print("Shape of X_poly : \n" , X_poly.shape)
```

```python
#comparing the values of X to X_poly
print("Entries of X : \n" , X[:5])
print("Entries of X_poly : \n" , X_poly[:5])
```

```python
print("Polynomial features names : \n" , poly.get_feature_names())
#get_feature_names can be used to get the value of all the features
```

```python
reg = LinearRegression().fit(X_poly , y)
line_poly = poly.transform(line)

plt.figure(figsize = (8 , 6))
plt.plot(line , reg.predict(line_poly) , 'r' , label= 'polynomial Regression')
plt.scatter(X , y , color = 'k')
plt.xlabel("features")
plt.ylabel("Target")
plt.legend()
plt.title("Polynomial regression")
#polynomial of high degree tend to behave very strangely at the extreme boundaries
```

```python
#Now we are going to use SVM to see how it is going to behave
from sklearn.svm import SVR

plt.figure(figsize = (8 , 6))
svr = SVR(gamma = 1).fit(X , y)
plt.plot(line , svr.predict(line) , 'g' , label = "Gamma 1" )

svr = SVR(gamma= 10).fit(X , y)
plt.plot(line , svr.predict(line) , 'r' , label = "Gamma 10")

plt.scatter(X , y , color = 'k')
plt.legend()
plt.xlabel("Feature")
plt.ylabel("Target")
plt.title("SVM with different gamma")
```

```python
#checking all this technique on boston housing data set
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

boston = load_boston()
X_train , X_test , y_train , y_test = train_test_split(boston.data , boston.target)

scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled= scaler.transform(X_test)
```

```python
poly = PolynomialFeatures(degree = 2).fit(X_train_scaled)
X_train_poly = poly.transform(X_train)
X_test_poly = poly.transform(X_test)
print("X_train shape : \n" , X_train.shape)
print("X_train_poly shape : \n" , X_train_poly.shape)
```

```python
#The new features represent all kind of 2 degree interaction between the polynomials
print("All the features are : \n" , poly.get_feature_names())
#The first new feature is a constant called 1
```

```python
from sklearn.linear_model import Ridge
ridge = Ridge().fit(X_train_scaled , y_train)
print("Training score : " , ridge.score(X_train_scaled ,  y_train))
print("Test scoer : " , ridge.score(X_test_scaled , y_test))
```

```python
#after interaction
ridge = Ridge().fit(X_train_poly , y_train)
print("Train accuracy : \n " , ridge.score(X_train_poly , y_train) )
print("test accuracy : \n" , ridge.score(X_test_poly , y_test))
#training accuracy is improved
```

```python
#using complex models like randomForest
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators = 100).fit(X_train_scaled , y_train)
print("test accuracy : " , rf.score(X_test_scaled , y_test))
rf = RandomForestRegressor(n_estimators = 100).fit(X_train_poly , y_train)
print("Test accuracy : " , rf.score(X_test_poly , y_test))
#there is only a slight increase in accuracy
#even without interaction random forest beats ridge
```

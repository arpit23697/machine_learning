```python
import mglearn
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Now we are going to tune the parameters of a model in the classical way
#We are going to tune the parameters of the SVC which are C and gamma
from sklearn.svm import SVC
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

iris = load_iris()
X_train , X_test , y_train , y_test = train_test_split(iris.data , iris.target , random_state =0)

print("Size of training set : " , X_train.shape)
print("Size of test set : " , X_test.shape)

best_score = 0
for gamma in [0.001 , 0.01 , 0.1 , 1 , 10 , 100]:
    for C in [0.001 , 0.01 , 0.1 , 1, 10 , 100]:
        model = SVC(C = C , gamma = gamma)
        model.fit(X_train , y_train)
        score = model.score(X_test , y_test)
        if (score > best_score):
            best_score = score
            best_parameters = {'C' : C , 'Gamma' : gamma}
            
print("Best score : " , best_score)
print("Best parameters : \n" , best_parameters)
```

```python
#We might be tempted to say that the accuracy is 97%
#but this not true bcoz paramters are now fit to the test set
#so we need to divide the data set into training set , test set and valid set
mglearn.plots.plot_threefold_split()
```

```python
#This is the implementation after splitting the data into three sets 
X_trainval , X_test , y_trainval , y_test = train_test_split (iris.data , iris.target , random_state = 0)
X_train, X_valid , y_train , y_valid = train_test_split(X_trainval , y_trainval , random_state = 1)

print("Size of training set is : " , X_train.shape)
print("Size of valid data set is : " , X_valid.shape)
print("Size of test set is : " , X_test.shape)

best_score = 0
for gamma in [0.001 , 0.01 , 0.1 , 1,10 , 100]:
    for C in [0.001 , 0.01 , 0.1 , 1 , 10 , 100]:
        model = SVC(C = C , gamma = gamma)
        model.fit(X_train , y_train)
        
        score = model.score(X_valid , y_valid )
        if (score > best_score):
            best_score = score
            best_parameters = {'C' : C  , 'gamma' : gamma}
            
print("Best score : " , best_score)
print("Best_paramters : " , best_parameters)


final_model = SVC(**best_parameters)
final_model.fit(X_trainval , y_trainval)
print("Test set score :" , final_model.score(X_test , y_test))
#any choices based on the test set accuracy , leaks information to the test set in the model
#we can use the cross validation to evaluate the performance 
#This is going to work betters
```

```python
from sklearn.model_selection import cross_val_score
best_score = 0
for gamma in [0.001 , 0.01 , 0.1 , 1, 10 , 100]:
    for C in [0.001 , 0.01 , 0.1 , 1, 10 ,100]:
        model = SVC(C = C , gamma = gamma)
        score = cross_val_score(model , X_trainval , y_trainval , cv = 5)
        
        score = score.mean()
        if (score > best_score):
            best_score = score
            best_parameters = {'C' : C , 'gamma' : gamma}
            
print("Best score : " , best_score)
print("Best paramters : \n" , best_parameters)

model = SVC(**best_parameters).fit(X_trainval , y_trainval)
print("Test score : " , model.score(X_test , y_test))
```

```python
#scikit learn gives the grid search method for tuning of the parameters
#implementing the model
param_grid = {'C' : [ 0.001 , 0.01 , 0.1 , 1 , 10 , 100],
              'gamma' : [0.001 , 0.01 , 0.1 , 1 , 10 , 100]}
print("Paramters grid : " , param_grid)
```

```python
#Now we implement the grid searcj method
from sklearn.model_selection import GridSearchCV
grid_search = GridSearchCV(SVC() , param_grid , cv = 5)
```

```python
#Note that the grid search will use the cross validation instead of the train_test_split
X_train, X_test , y_train , y_test = train_test_split(iris.data , iris.target , random_state = 0)
grid_search.fit(X_train , y_train)
print("Test score is : " , grid_search.score(X_test , y_test))
#grid_search finds the best model and then fit the data onto the model
```

```python
print("Best parameters : " , grid_search.best_params_)
```

```python
print("Best cross validation score : " , grid_search.best_score_)    #this is the mean score
```

```python
print("Best estimators : " , grid_search.best_estimator_)
#this is going to give you the best model 
```

```python

def plot_heatmap (scores , gamma , C):
#using the hear mop to visualise it
#print(scores)


    fig , ax = plt.subplots(figsize = (6, 4.8))
    im = ax.pcolor(scores)
    ax.set_xticks(np.arange(len(gamma)) + 0.5)
    ax.set_yticks(np.arange(len(C)) + 0.5)
    ax.set_xticklabels(gamma)
    ax.set_yticklabels(C)
    ax.set_xlabel("gamma")
    ax.set_ylabel("C")
    for i in range(len(gamma)):
        for j in range(len(C)):
            string = "{:.2f}".format(scores[j , i])
            text = ax.text(i+0.5 , j+0.5 , string  , weight = "bold" , ha = 'center' , va = 'center')

    ax.set_title("Analysis of all the models")
    fig.tight_layout()
    #here the color is used to encode the cross validation accuracy 
    #It is very clear that the SVC is very sensitive to the setting of the parameter
    #it is important to note that the optimum value should not be on the edges of the plot
```

```python
#now we analyse the result of the grid search
#it is good to visualize the results of the cross validation to understand how the model generalizes
#we can find all the aspects of the grid search in cv_results_ attribute
#it returns a dictionary
#print(grid_search.cv_results_)
results = pd.DataFrame(grid_search.cv_results_)
print(results.head())
scores = np.array(results.mean_test_score).reshape(6, 6)
gamma = [0.001 , 0.01 , 0.1 , 1 , 10 , 100]
C = [.001 , 0.01 , 0.1 , 1 , 10 , 100]
```

```python
plot_heatmap (scores , gamma , C)
```

```python
#some results where the result is less ideal because the search ranges are not properly
param_grid_linear = {'C' : np.linspace(1 , 2 , 6),
                    'gamma' : np.linspace (1 , 2, 6)}
C = np.linspace(1 , 2 , 6)
gamma = np.linspace(1 , 2 , 6)
grid_search = GridSearchCV(SVC() , param_grid_linear , cv= 5)
grid_search.fit(X_train , y_train)
print("Test score : " , grid_search.score(X_test , y_test) )
print("Best parameters : " , grid_search.best_params_)
results = pd.DataFrame(grid_search.cv_results_)
scores = np.array(results.mean_test_score).reshape(6 , 6)
plot_heatmap(scores , gamma , C)
```

```python
import mglearn
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#now importing the wave dataset
X , y = mglearn.datasets.make_wave(n_samples = 40)
```

```python
print("Shape of X : \n" , X.shape)
print("Shape of y : \n" , y.shape)
```

```python
plt.plot(X , y , 'bo')
plt.xlabel('feature')
plt.ylabel("target")
plt.title("data points")
```

```python
from sklearn.model_selection import train_test_split
X_train ,X_test , y_train, y_test = train_test_split(X , y , random_state = 42)
```

```python
#now using the model and fitting it
#simplest linear regression model (with no regularisation)
from sklearn.linear_model import LinearRegression
lr = LinearRegression().fit(X_train , y_train)
```

```python
#printing out the parameters
#coef_ this is the attribute used to store the coefficients ; this is the array
#intercept_ is the attrbute used to store the intercept
print("lr.coef_ : ", lr.coef_)
print("lr.intercept_ : ", lr.intercept_)
```

```python
#plotting the graph
plt.figure(figsize = (10 , 7))

plt.plot(X, y , 'bo')
plt.title('Data points and prediction')
plt.xlabel("Feature")
plt.ylabel("Target")

line = np.linspace (-3 , 3 , 1000)
y_pred = line*lr.coef_[0] + lr.intercept_
plt.plot(line , y_pred)
plt.grid()
```

```python
#printing the accuracies
print("Training set accuracy : " , lr.score(X_train , y_train))
print("Test set accuracy : " , lr.score(X_test , y_test))
```

```python
#importing the boston dataset
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
print("Keys of the boston dataset are : " , boston.keys())
```

```python
print("Shape of data is : ", boston.data.shape)
print("Shape of target is : " , boston.target.shape)
```

```python
#this is to feature engineer the features and include all the degree = 2 features
from sklearn.preprocessing import MinMaxScaler , PolynomialFeatures
X = boston.data
y = boston.target
X = MinMaxScaler().fit_transform(boston.data)
X = PolynomialFeatures (degree = 2 , include_bias = False).fit_transform(X)
X.shape
```

```python
#splitting the dataset
X_train , X_test, y_train , y_test = train_test_split(X , y)
print("Shape of X_train is : " , X_train.shape)
print("Shape of X_test is : " , X_test.shape)
```

```python
#fitting the model
lr = LinearRegression().fit(X_train , y_train)
```

```python
#printing the accuracies
print("Train set accuracy : " , lr.score(X_train , y_train))
print("Test set accuracy : " , lr.score(X_test , y_test) )
```

```python
#next model is ridge regression
#this is linearRegression with regularisation
from sklearn.linear_model import Ridge
```

```python
ridge = Ridge().fit(X_train , y_train)
```

```python
#printing the accuracies
print("Training set accuracies : " , ridge.score(X_train , y_train))
print("Test set accuracies : " , ridge.score(X_test , y_test))
```

```python
#visualising the effect of the regualisation parameter
training = []
test = []
lamdas = np.linspace (0.1 , 10 , 20)
for alphas in lamdas:
    ridge = Ridge(alpha = alphas).fit(X_train , y_train)
    training.append(ridge.score(X_train , y_train))
    test.append(ridge.score(X_test , y_test))
```

```python
plt.plot(lamdas , training , label = 'training')
plt.plot(lamdas , test , label = 'test')
plt.xlabel('lamdas')
plt.ylabel('accuracy')
plt.legend()
plt.title('analysis')
#training accuracy is always higher than the test set accuracy
#when lamda is very small , the model is overfitting 
#when lamda is high , the model is underfitting , test and traing accuracies both are low
```

```python
#how the coefficients vary with the increase in the value of alpha
ridge1 = Ridge(alpha = 1).fit(X_train , y_train)
ridge10 = Ridge(alpha = 10).fit(X_train , y_train)
ridge01 = Ridge(alpha = .1).fit(X_train , y_train)

plt.figure(figsize = (10 , 7))

plt.plot(ridge1.coef_ , '^b' , label = 'alpha = 1')
plt.plot(ridge10.coef_ , '^g' , label = 'alpha = 10')
plt.plot(ridge01.coef_ , '^r' , label = 'alpha = 0.1')

plt.plot(lr.coef_ , 'o' , label = 'alpha = 0 ')
plt.legend()
plt.xlabel('Feature')
plt.ylabel('coefficient magnitude')
plt.ylim(-25 , 25)

#when alpha is 0.1 , the coefficients are varying largely
#when alpha = 0 , linear regression , the points are varying very widely , some going out of the figure
#when alpha = 10 , the points are restricted close to 0
```

```python
#now seeing the effect of training set size with the increase value of training set size
#as the training set size increase , the accuracies fall down , 
#as the size increase , test set accuracy increase and training set accuracy decrease
#regression model is generally better then simple linear regression model
totalSize = X.shape[0]
samples = np.linspace (60 , totalSize, 10)
training_ridge = []
training_ls = []
test_ridge = []
test_ls = []
for n in samples:
    X1_ , X2_ , y1_ , y2_ = train_test_split(X[:int(n) , :] , y[ : int(n)])
    ridge = Ridge(alpha = 1.0).fit(X1_ , y1_)
    lr = LinearRegression().fit(X1_ , y1_)
    training_ridge.append(ridge.score(X1_ , y1_))
    training_ls.append(lr.score(X1_ , y1_))
    test_ridge.append(ridge.score(X2_ , y2_))
    test_ls.append(lr.score(X2_ , y2_))
```

```python
#plotting the graph
plt.figure(figsize = (10, 7))
plt.plot( samples, training_ridge, label = 'training_ridge')
plt.plot(samples , training_ls , label = 'training_lr')
plt.plot(samples , test_ridge ,'--'  ,label = 'test_ridge')
plt.plot(samples , test_ls, '--'  , label = 'test_lr' )
plt.legend()
plt.xlabel('sample_size')
plt.ylabel('accuracy')
plt.title('accuracy Vs Sample_size')
plt.ylim(0 , 1.2)

#these are the learning curves
#since there is no regularisation , the training accuracy of the regularised model is less than the linera regression model
#regularised model are better since there training and test set accuracies are good, almost equal
#the linear regression model tends to overfit
#with enough training data , regularisation becomes less important, lr and ridge perform almost the same
#regularised model are good, since there test accuracies are good as compared to linear regression test set accuracies
#if more data is available , it becomes harder to fit depicted by decrease in training set accuracy of lr
```

```python
#Now shows the effect of the lasso regression
#It is the same as regularised regression but in this some of the coefficients are exactly zero
#this means some features are entirely ignored by the lasso
#making some of the features zero , reveal the most important features of your model
from sklearn.linear_model import Lasso
lasso = Lasso().fit(X_train , y_train)
```

```python
#printing the accuracies
print("Training set accuracy : " , lasso.score(X_train , y_train))
print("Test set accuracy : ", lasso.score(X_test , y_test))
print("Number of feature used : " , np.sum(lasso.coef_ != 0))
#this shows very less number of feature are used 
#so the training and the test accuracies are very small
#This shows that we are underfitting 
#the model is too simple as it is taking into account only two features
```

```python
#decreasing the value of alpha
#so also increasing the max_iter
#this is fitting a more complex model , so the better training accuracy
lasso001 = Lasso(alpha = 0.01 , max_iter = 10000).fit(X_train , y_train)
print("Training accuracy is : " , lasso001.score(X_train , y_train))
print("Test accuracy is : " , lasso001.score(X_test , y_test))
print("Number of feature used : ", np.sum(lasso001.coef_ != 0))
```

```python
#if we set alpha too low , we reduce the effect of regularisation to great extent 
#so overfitting the model
lasso0001 = Lasso(alpha = 0.0001 , max_iter = 100000).fit(X_train , y_train)
print("Training accuracy is : " , lasso0001.score(X_train , y_train))
print("Test accuracy is : " , lasso0001.score(X_test , y_test))
print("Number of feature used : ", np.sum(lasso0001.coef_ != 0))
```

```python
#plotting the coefficients of different models
plt.figure(figsize = (10 , 8))
plt.plot(lasso.coef_ , 's' , label = 'alpha = 1.0')
plt.plot(lasso001.coef_ , '^' , label = 'alpha = 0.01')
plt.plot(lasso0001.coef_ , '^' , label = 'alpha = 0.0001')

plt.plot(ridge01.coef_ , '^' , label = 'ridge model')

plt.legend()
plt.xlabel('Feature coefficient')
plt.ylabel('Coeffient magnitude')
plt.title('Analysis')
plt.ylim(-20 , 20)

#corresponding to alpha = 0.0001 model is separated
#corresponding to alpaha = 1.0 , most of the features are zero
#alpha = 0.01 have the same performance as alpha = 0.1 for ridge model.
# but in case of ridge model, all coeff  are non zero
```

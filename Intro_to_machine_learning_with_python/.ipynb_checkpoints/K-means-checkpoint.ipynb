```python
def discrete_scatter (X , y , mark , classes , markerWidth = 1 , markerSize = 60):
    colors = ['red' , 'yellow', 'blue' , 'green' , 'purple' , 'brown' , 'black' , 'gray' , 'olive' , 'cyan']
    for i in range(classes):
        temp = X[y == i]
        plt.scatter(temp[: , 0] , temp[:, 1] , label = "Class " + str(i) , marker = mark,
                   linewidth = markerWidth , edgecolor = 'black' , s = markerSize , color = colors[i])
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
```

```python
from sklearn.datasets import make_blobs
X , y = make_blobs(random_state = 1)
```

```python
discrete_scatter(X , y , 'o' , 3 , 1 , 60)
plt.legend()
plt.title('data points')
```

```python
#build the clustering model
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 3)
kmeans.fit(X)
```

```python
#The labels of the data points of X are 
print("Cluster membership : \n" , kmeans.labels_)
#this show the label point of all the data points
```

```python
#assiginng cluster labels to the new points without changing the data
#Note the existing model is not changed
#running predict on the training set returns the same result
print("Predicted labels : \n" , kmeans.predict(X))
```

```python
#plotting the result of the kmeans on this dataset
#def discrete_scatter (X , y , mark , classes , markerSize)
plt.figure(figsize = (8 , 6))
discrete_scatter(X , kmeans.labels_ , 'o' , 3 , 1,100 )
discrete_scatter(kmeans.cluster_centers_ , np.array([0 , 1 , 2]) , '^' , 3 , 3 , 200)
plt.title("Result of kmeans algorithm")
```

```python
#using fewer cluster centers
kmeans = KMeans(n_clusters = 2)
kmeans.fit(X)

plt.figure(figsize = (8 , 6))
discrete_scatter(X , kmeans.labels_ , 'o' , 2 , 1 , 100)
discrete_scatter(kmeans.cluster_centers_ , np.array([0 , 1]) , '^' , 2 , 3, 200)
plt.title("Plot with 2 clusters")
```

```python
#making 5 clusters
kmeans = KMeans(n_clusters = 5)
kmeans.fit(X)
plt.figure(figsize = (8 , 6))
discrete_scatter(X , kmeans.labels_ , 'o' , 5 , 1 , 100)
discrete_scatter(kmeans.cluster_centers_  , np.array([0 , 1, 2, 3 , 4])  , '^' , 5, 3 , 200)
plt.title("Plot with 5 clusters")
```

```python
#Even if you know the correct number of clusters , kmeans may not be able to extract it properly always
#kmeans can only capture only relatively simple shape
#kmeans also assumes that all the clusters have the same "diameter" because it draws boundary to be exactly in between
X_varied , y_varied = make_blobs(n_samples = 200 , cluster_std = [1.0 , 2.0 , 0.5] , random_state = 170)
plt.figure(figsize = (8 , 6))
discrete_scatter(X_varied , y_varied , 'o' , 3 , 1 , 100)
plt.title("Data points for varied data set")
```

```python
kmeans = KMeans (n_clusters = 3)
kmeans.fit(X_varied)
plt.figure(figsize = (8 , 6))
discrete_scatter(X_varied , kmeans.labels_ , 'o' , 3 , 1 , 100)
discrete_scatter(kmeans.cluster_centers_ , np.array([0 , 1, 2]) , '^' , 3 , 2 , 200)
plt.title("Kmeans for varied data set")

#This is the result of the kmeans algorithm
#It is not consistent , as the clustering is not good
#The result is slightly mixed up
```

```python
#kmeans also assumes that all the directions are important
#So varied data set in the diagonal direction is not good for kmeans
X , y = make_blobs (random_state = 170 , n_samples = 600)
rnd = np.random.RandomState(74)

transformation = rnd.normal(size= (2 , 2))
X = np.dot (X , transformation)

#plotting the original data
plt.figure(figsize = (8 , 6))
discrete_scatter (X , y , 'o' , 3 , 1 , 100)
plt.title("Original data points")
```

```python
#applying kmeans and seeing the result
kmeans = KMeans(n_clusters = 3)
kmeans.fit(X)

#plotting the result
plt.figure(figsize = (8 , 6))
discrete_scatter (X , kmeans.labels_ , 'o' , 3 , 1, 100)
discrete_scatter(kmeans.cluster_centers_ , np.array([0 , 1, 2]) , '^' , 3, 2 , 200)
plt.title("Kmeans result")
```

```python
#if the clusters have more complex shape , then kmeans do not preform very well
from sklearn.datasets import make_moons
X , y = make_moons (n_samples = 200 , noise = 0.05  , random_state= 0)

plt.figure(figsize = (8 , 6))
discrete_scatter(X , y , 'o' , 2 , 1 , 100)
plt.title("Original data points")
```

```python
kmeans = KMeans (n_clusters = 2)
kmeans.fit(X)

plt.figure(figsize = (8 , 6))
discrete_scatter(X , kmeans.labels_ , 'o' , 2 , 1 , 100)
discrete_scatter(kmeans.cluster_centers_ , np.array([0 ,1]) , '^' , 2 , 2 , 300)
plt.title("Result of kmeans")
```

```python
#Kmeans , PCA and nmf have intresting parallels
#PCA tries to find the direction of maximum variance
#nmf tries to find additive components
#Both method tries to find the new representation of how to express the data point as the sum of components
#kmeans tries to represent the data point with the help of cluster centers
#This view , where each data point is represented as the cluster center is known as vector quantization
#Now lets visualize how each method is working in representation
from sklearn.datasets import fetch_lfw_people
people = fetch_lfw_people(min_faces_per_person = 20 , resize = 0.7)
people.keys()
print ("Size of people.data : " , people.data.shape) #there are 3023 data points , with each having 5655 features
print("Shape of image is : " , people.images[0].shape)
```

```python
#Now applying different methods to represent the data
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(people.data , people.target)
```

```python
from sklearn.decomposition import NMF
from sklearn.decomposition import PCA
```

```python
nmf = NMF(n_components = 100 , random_state= 0)
nmf.fit(X_train)
pca = PCA (n_components = 100 , random_state= 0)
pca.fit(X_train)
kmeans = KMeans (n_clusters = 100 , random_state = 0)
kmeans.fit(X_train)

X_reconstructed_pca = pca.inverse_transform(pca.transform(X_test))
X_reconstructed_nmf = nmf.inverse_transform(nmf.transform(X_test))
X_reconstructed_kmeans = kmeans.cluster_centers_[kmeans.predict(X_test)]
```

```python
#plotting the components as well
image_shape = people.images[0].shape
fig , axes = plt.subplots(3 , 5 , figsize = (12 ,12) ,
                         subplot_kw = {'xticks' : () , 'yticks' : ()})

fig.suptitle("Components")
for i , ax , image in zip (range(5) ,axes[0 , : ] , pca.components_):
    ax.imshow(image.reshape(image_shape) , cmap = 'gray')
    ax.set_title("Pca component " + str(i))
    
for i , ax , image in zip (range(5) ,axes[1 , : ] , nmf.components_):
    ax.imshow(image.reshape(image_shape) , cmap = 'gray')
    ax.set_title("nmf component " + str(i))
    
for i , ax , image in zip (range(5) ,axes[2 , : ] , kmeans.cluster_centers_):
    ax.imshow(image.reshape(image_shape) , cmap = 'gray')
    ax.set_title("kmeans cluster " + str(i))
```

```python
#now plotting each of the images for the visualisation purpose

fig , axes = plt.subplots(4 , 5 , figsize = (12, 12) , 
                         subplot_kw = {'xticks' : () , 'yticks' : ()})
fig.suptitle("Extracted components")
for ax , image in zip (axes[0 , :] , X_test):
    ax.imshow(image.reshape(image_shape) , cmap = 'gray')
    ax.set_title('Original')


for ax , image in zip (axes[1 , :] , X_reconstructed_pca):
    ax.imshow(image.reshape(image_shape) , cmap = 'gray')
    ax.set_title('PCA')
    
for ax, image in zip (axes[2 , : ] , X_reconstructed_nmf):
    ax.imshow(image.reshape(image_shape) , cmap = 'gray')
    ax.set_title('NMF')
    
for ax , image in zip(axes[3 , : ] , X_reconstructed_kmeans):
    ax.imshow(image.reshape(image_shape) , cmap = 'gray')
    ax.set_title('KMeans')
```

```python
#one intresting aspect of vector quantization is using k-means, is that we can use many more clusters to represent our data well
#for example , with more cluster points we would be able to find more expressive representation of moon dataset
X , y  = make_moons (n_samples = 200 , noise = 0.05 , random_state = 0)
kmeans = KMeans (n_clusters = 10)
kmeans.fit(X)

plt.figure (figsize = (8, 6))
discrete_scatter (X , kmeans.labels_ , 'o' , 10 , 1 , 100) 
discrete_scatter(kmeans.cluster_centers_ , np.arange(10) , '^' , 10 , 2 , 200)
plt.title("Kmeans with 10 cluster points")
```

```python
print("Cluster membership : ", kmeans.labels_ )
```

```python
#this in some sense is a new representation with 10 features
#all feature points being 0 apart from the one to which the data point is assigned to
#using this representation it is possible to separate it using the linear data set
#we can get a more expressive representation , by looking the distance of the data point to each of the clusters
distance_features = kmeans.transform(X)
print("Distance features shape : " , distance_features.shape)
print("Distance features : \n" , distance_features)
```

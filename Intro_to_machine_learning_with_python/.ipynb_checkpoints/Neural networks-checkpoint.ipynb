```python
import mglearn
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Non linear function used in MLP neural network
line = np.linspace (-3 , 3 ,100)
plt.plot(line , np.tanh(line) , label = 'tanh')
plt.plot(line , np.maximum(line , 0) , label = 'relu')
plt.legend(loc = 'best')
plt.xlabel("x")
plt.ylabel("relu(x) , tanh(x)")
```

```python
#plotting the datap point
def plot_data_points(X , y):
    class0 = np.ones((0 , 2))
    class1 = np.ones((0 , 2))
    for indices in range(X.shape[0]):
        if (y[indices] == 0):
            class0 = np.vstack((class0 , np.array(X[indices])))
        else:
            class1 = np.vstack((class1 , np.array(X[indices])))
    plt.plot(class0[: , 0] , class0[: , 1] , 'o' , label = "class 0")
    plt.plot(class1[: , 0] , class1[: , 1] , 'r^' , label = "class 1")
    plt.xlabel('Feature 1')
    plt.ylabel("feature 2")
```

```python
def plot_decision_boundary(model , X , y):
    eps = 0.5
    x_min , x_max = X[: , 0].min() - eps , X[: , 0].max() + eps
    y_min , y_max = X[: , 1].min() - eps , X[: , 1].max() + eps
    
    xx = np.linspace(x_min , x_max , 100)
    yy = np.linspace(y_min , y_max , 100)
    
    X1 , X2 = np.meshgrid(xx , yy)
    X_grid = np.c_[X1.ravel() , X2.ravel()]
    
    Z = model.predict_proba(X_grid)[: , 1]
    plt.contourf(X1 , X2 , Z.reshape(X1.shape) , alpha = 0.5 , levels = [0 , 0.5 , 1])
```

```python
#using the two moon dataset
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
X , y = make_moons(n_samples = 100 , noise = 0.25 , random_state = 3)
X_train , X_test , y_train , y_test = train_test_split(X , y , stratify = y , random_state = 42)
```

```python
#plotting the data points
plot_data_points(X , y)
plt.legend()
plt.title("Data points")
```

```python
#fitting the model and plotting the curve
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(solver='lbfgs', random_state=0).fit(X_train , y_train)
```

```python
plot_decision_boundary(mlp , X , y)
plot_data_points(X , y)
plt.legend(loc = 'best')
plt.title('Decision boundary')
```

```python
#decreasing the number of hidden units
mlp2 = MLPClassifier(solver = 'lbfgs' , random_state = 0 , hidden_layer_sizes = [10])
mlp2.fit(X_train , y_train)
#plotting the decision boundary
plot_decision_boundary(mlp2 , X , y)
plot_data_points(X , y)
plt.legend(loc = 'best')
plt.title("Decision boundary")

#this is still giving a pretty good decision boundary with just 10 hidden units
#The default non linearity is relu
#With 10 hidden units , it will make a decision boundary with atmost 10 straigh lines
```

```python
#neural network with 2 hidden layers , 10 units each
mlp3 = MLPClassifier(solver = 'lbfgs' , random_state = 0 , hidden_layer_sizes = [10,10])
mlp3.fit(X_train , y_train)

#plotting the decision boundary
plot_decision_boundary(mlp3 , X , y)
plot_data_points(X , y)
plt.legend()
plt.title("Decision boundary")
```

```python
#neural network with 2 hidden layers , 10 units each and tanh activation
mlp4 = MLPClassifier(solver = 'lbfgs' , random_state = 0 , activation = 'tanh' , hidden_layer_sizes = [10,10])
mlp4.fit(X_train , y_train)

#plotting the decision boundary
plot_decision_boundary(mlp4 , X , y)
plot_data_points(X , y)
plt.legend()
plt.title("Decision boundary")
```

```python
#Another way to control the complexity of the model is by adjusting the parameter alpha (regularisation) 
#below shows the examples of different decision boundaries 
mlp = MLPClassifier(solver = 'lbfgs' , random_state = 0 , alpha = 0.1).fit(X_train , y_train)
#plotting the decision boundary
plot_decision_boundary(mlp , X , y)
plot_data_points(X , y)
plt.legend()
plt.title("Decision boundary")
```

```python
mlp = MLPClassifier(solver = 'lbfgs' , random_state = 0 , alpha = 1).fit(X_train , y_train)
#plotting the decision boundary
plot_decision_boundary(mlp , X , y)
plot_data_points(X , y)
plt.legend()
plt.title("Decision boundary")
```

```python
mlp = MLPClassifier(solver = 'lbfgs' , random_state = 0 , alpha = 10).fit(X_train , y_train)
#plotting the decision boundary
plot_decision_boundary(mlp , X , y)
plot_data_points(X , y)
plt.legend()
plt.title("Decision boundary")
```

```python
mlp = MLPClassifier(solver = 'lbfgs' , random_state = 0 , alpha = 10 , hidden_layer_sizes = [100]).fit(X_train , y_train)
#plotting the decision boundary
plot_decision_boundary(mlp , X , y)
plot_data_points(X , y)
plt.legend()
plt.title("Decision boundary")
```

```python
#note that the parameters of the neural network are randomly initialised
#so using different value of random-seed may give different results
#For large and complex model it is not going to effect much
#But for small model it should be kept in mind
mlp = MLPClassifier(solver = 'lbfgs' , random_state = 0 ,).fit(X_train , y_train)
#plotting the decision boundary
plot_decision_boundary(mlp , X , y)
plot_data_points(X , y)
plt.legend()
plt.title("Decision boundary")
```

```python
mlp = MLPClassifier(solver = 'lbfgs' , random_state = 1).fit(X_train , y_train)
#plotting the decision boundary
plot_decision_boundary(mlp , X , y)
plot_data_points(X , y)
plt.legend()
plt.title("Decision boundary")
```

```python
#using the neural network on breast cancet dataset
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
X_train , X_test , y_train , y_test = train_test_split(cancer.data , cancer.target , random_state = 0) 
```

```python
#fitting the model
mlp = MLPClassifier( random_state = 0)
mlp.fit(X_train , y_train)
print("Train_accuracy : " , mlp.score(X_train , y_train))
print("Test accuracy : " , mlp.score(X_test , y_test))
#the accuracy of the model is very low 
#So to improve the performance feature scale the model
```

```python
#scaling the data for better functioning of the data
mean_on_train = X_train.mean(axis = 0)
std_on_train = X_train.std(axis = 0)
X_train_scaled = (X_train - mean_on_train) / std_on_train
X_test_scaled = (X_test - mean_on_train) / std_on_train
```

```python
mlp = MLPClassifier(random_state = 0)
mlp.fit(X_train_scaled , y_train)
print("Train accuracy : " , mlp.score(X_train_scaled , y_train))
print("Test accuracy :" , mlp.score(X_test_scaled , y_test))
#The model is improved 
#But we are getting the warning to increase the number of iterations , so as to converge properly
```

```python
#increasing the number of iterations
mlp = MLPClassifier (random_state = 0 , max_iter = 1000)
mlp.fit(X_train_scaled , y_train)
print("Train accuracy : " , mlp.score(X_train_scaled , y_train))
print("Test accuracy : " , mlp.score(X_test_scaled , y_test))
#increasing the number of iterations increases the train and test accuracy
```

```python
#increasing the number of iterations ,and increasing the value of alpha from 0.0001 to 1.0
mlp = MLPClassifier (random_state = 0 , max_iter = 1000 , alpha = 1)
mlp.fit(X_train_scaled , y_train)
print("Train accuracy : " , mlp.score(X_train_scaled , y_train))
print("Test accuracy : " , mlp.score(X_test_scaled , y_test))
#increasing the number of iterations increases the train and test accuracy
```

```python
#visualising the weights of the neural network
plt.figure(figsize = (20 , 7))
plt.imshow(mlp.coefs_[0] , interpolation = 'none' , cmap = 'viridis')
plt.yticks(range(30) , cancer.feature_names)
plt.xlabel("Columns in weight matrix")
plt.ylabel("Input feature")
plt.colorbar()
```
